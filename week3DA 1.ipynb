QUESTION2 For each of the numeric attributes (proteins upto vitamins) , identify and replace all missing data(indicated with -1) with the arithmetic mean of the attribute.


```python
cereals["calories"].replace({-1:cereals["calories"].mean()}, inplace=True)
cereals["protein"].replace({-1:cereals["protein"].mean()}, inplace=True)
cereals["fat"].replace({-1:cereals["fat"].mean()}, inplace=True)
cereals["sodium"].replace({-1:cereals["sodium"].mean()}, inplace=True)
cereals["fiber"].replace({-1:cereals["fiber"].mean()}, inplace=True)
cereals["carbo"].replace({-1:cereals["carbo"].mean()}, inplace=True)
cereals["sugars"].replace({-1:cereals["sugars"].mean()}, inplace=True)
cereals["potass"].replace({-1:cereals["potass"].mean()}, inplace=True)
cereals["vitamins"].replace({-1:cereals["vitamins"].mean()}, inplace=True)
new_cereals=cereals
new_cereals
```

QUESTION4 For each of the numeric attributes (proteins upto vitamins), identify and replace all noisy data with the median of attribute. 

```python
numeric = ["calories", "protein", "fat" , "sodium", "fiber" , "carbo", "sugars" ,"potass", "vitamins" ]

for i in numeric:
  inter_quartile =new_cereals[i].quantile(0.75) - new_cereals[i].quantile(0.25)
  condition = (new_cereals[i]>=new_cereals[i].quantile(0.75)+inter_quartile*1.5) | (new_cereals[i]<=new_cereals[i].quantile(0.25)-inter_quartile*1.5)
  new_cereals.loc[condition,i] = new_cereals[i].median()
b=new_cereals.describe()
b


```

QUESTION15 Design a Linear regression model to predict the rating of a cereal based on top 3 related nutrients. Tabulate the accuracy of the model using a 70 ,30 split.

```python
new_df = new_cereals[['calories','protein','carbo','rating']]
new_df
```

```python
from sklearn.model_selection import train_test_split
X = new_df[['calories','protein','carbo']]
y = new_df['rating']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
y_pred =reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred, squared=False)
```
